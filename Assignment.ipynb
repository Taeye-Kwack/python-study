```python
# Assignment 8.4

fname = input("Enter file name: ")
fh = open(fname)
lst = list()
for line in fh:
    line = line.rstrip()
    line = line.split()
    for piece in line:
        if not piece in lst : lst.append(piece)

lst = sorted(lst)
print(lst)
```

```python
# Assignment 8.5

fname = input("Enter file name: ")
if len(fname) < 1:
    fname = "mbox-short.txt"
    
fh = open(fname)

count = 0
for line in fh :
    line = line.rstrip()
    if not line.startswith("From ") : continue
    words = line.split()
    word = words[1]
    count += 1
    print(word)

print("There were", count, "lines in the file with From as the first word")
```

```python
# Assignment 7.2

# Use the file name mbox-short.txt as the file name
fname = input("Enter file name: ")
try : fh = open(fname)
except : 
    print("Error")
    quit()

total, count = 0, 0
for line in fh:
    if not line.startswith("X-DSPAM-Confidence:"):
        continue
    else : 
        count += 1
        idx = line.find(":")
        num = line[idx+1:]
        num = num.strip()
        num = float(num)
        total += num
        #print(line)
avg = total / count
print("Average spam confidence:", avg)
#print("Done")
```

```python
# Assignment 9.4

name = input("Enter file:")
if len(name) < 1:
    name = "mbox-short.txt"
handle = open(name)

x = dict()
for line in handle :
    line.rstrip()
    if not line.startswith("From ") : continue
    email = line.split()[1]
    x[email] = x.get(email,0) + 1
    
bigcount, bigemail = None, None
for word, count in x.items():
    if bigcount is None or count > bigcount :
        bigcount = count
        bigemail = word
 
print(bigemail, bigcount)
```
