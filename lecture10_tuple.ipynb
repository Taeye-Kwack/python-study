```python
# Tuples are comparable. Compare from the first components
(0,1,2) < (5,1,2) #True
(0,1,2000) < (0,3,4) #True 
```

```python
(x,y) = (4,'fred')
print(y) # assignment statement

# sorted with the value
c = {'a':10, 'b':1, 'c':22}
tmp = list()

for k,v in c.items(): 
    tmp.append((v,k))
print(tmp)

tmp = sorted(tmp, reverse = True)
print(tmp)

sorted([(v,k) for k,v in c.items()]) # same as tmp; list comprehension
```

```python
# Assignment 10.2

name = input("Enter file:")
if len(name) < 1:
    name = "mbox-short.txt"
handle = open(name)

counts = dict()
for line in handle:
    if not line.startswith("From ") : continue
    time = line.split()[5]
    time = time.split(":")[0]
    counts[time] = counts.get(time,0) + 1

counts = sorted([(v,k) for (v,k) in counts.items()])

for v,k in counts:
    print(v,k)
```
